```python
import numpy as np
import pandas as pd
import seaborn as sb
%matplotlib inline
import matplotlib.pyplot as plt
import math
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/ArifHidaya123/datamining/gh-pages/exams.csv")
print(data)
```

```python
data
```

```python
def dt1(i,j):
  P = 9
  Q = 0
  for col in ["gender", "lunch", "test_prep_course", "math"]:
    if data[col][i] == data[col][j]:
      Q += 1
  return (P-Q)/P
```

```python
print(dt1(1,2))
print(dt1(1,3))
print(dt1(1,4))
```

```python
def dt2(i,j,h):
  result = 0
  for col in ["math"]:
    result = math.pow(abs(data[col][i] - data[col][j]), h)
  return result**(1/h)
```

```python
print(dt2(1,2,1))
print(dt2(1,3,1))
print(dt2(1,4,1))
```
